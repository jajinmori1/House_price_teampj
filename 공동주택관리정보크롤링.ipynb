### 공동주택관리시스템 크롤링

```python
# ChromeDriver 자동 설치 모듈
import chromedriver_autoinstaller
# Chrome을 제어하기 위한 객체
from selenium import webdriver
# Chrome이 웹 페이지 로딩을 완료 할 때까지 최대 n초간 대기하는 기능.
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
from pandas import DataFrame
# 파이썬 프로그램에 지정된 시간동안 랙을 거는 기능을 위해 사용
import time
#경우의 수
import itertools
import openpyxl
```

```python
# 크롬드라이버 자동 설치
chromedriver_autoinstaller.install()
df = DataFrame()
```

```python
err_flag = 1
```

```python
# 크롬드라이버를 통해 크롬을 실행시킴
# -> driver 객체는 Chrome 자체

driver = webdriver.Chrome()
# 크롬브라우저가 준비될 때 까지 최대 5초씩 대기

driver.implicitly_wait(5)

driver.get("http://www.k-apt.go.kr/kaptinfo/openkaptinfo.do")
    # 페이지를 이동하는 동안 최대 2초간 대기
time.sleep(2)
select_sido = Select(driver.find_element_by_css_selector("select[class='combo_SIDO']"))
select_sido.select_by_index(1)
time.sleep(2)

select_sgg = Select(driver.find_element_by_css_selector("select[class='combo_SGG']"))
sgg_index = 5 #시작 시군구 상자(A:1 , B:6, C:11, D:16, E:21) 로봇인증시 수정

before = ""
# ---------------------------------시군구 선택상자-----------------------------------------
while(True):
    
    select_sgg.select_by_index(sgg_index)
    time.sleep(1)
    select_emd = Select(driver.find_element_by_css_selector("select[class='combo_EMD']"))
    
    if err_flag == 1:
        emd_index = 3 # 로봇인증시 수정
    else:
        emd_index = 1
    # ---------------------------------읍면동 선택상자-----------------------------------------
    while(True):

        try:
            select_emd.select_by_index(emd_index)
            try:
                alert = Alert(driver) # 읍면동에 단지가 없으면 다음 읍면동 선택
                alert.accept()
                emd_index += 1
                continue
            except:
                'No Alert'
        except:
            break # 읍면동 인덱스가 벗어나면(에러발생) 시군구 선택으로 이동
        time.sleep(2)
        
        if err_flag == 1:
            danji_index = 19 # 로봇인증시 수정
            err_flag = 0
        else:
            danji_index = 1
        
        # ---------------------------------데이터 수집-----------------------------------------
        while(True):
            try:
                button = WebDriverWait(driver, 3).until(lambda x: x.find_element_by_css_selector("tr[id='%d']" %danji_index))
                button.click()
                time.sleep(1)
            except:
                print('단지수 : %d ***수가 맞지 않으면 확인***' %(danji_index - 1))
                break # 단지 인덱스가 벗어나면(에러발생) 읍면동 선택으로 이동

            button = WebDriverWait(driver, 3).until(
                    lambda x: x.find_element_by_css_selector("img[src='/images/sub/tab_kaptinfo5_out.png']"))
            button.click()
            time.sleep(1)

            html = driver.page_source
            soup = BeautifulSoup(html, "html.parser")

            id = soup.find_all('td',id=True)

            id_list = []
            for it in id:
            # 각 원소(링크)에 속성들(attrs) 중에 id 속성이 있다면 그 값을 별도로 준비한 리스트에 추가
                if "id" in it.attrs:
                    id_list.append(it['id'])

            name = soup.find_all('th',class_='type')

            컬럼명_list = []

            for i in name:
                k = i.text.strip()
                컬럼명_list.append(k)

            tmp = []
            data_list = []

            for i,v in enumerate(컬럼명_list) :
                data = soup.select("#%s"%id_list[i])
                tmp.append(data)

            for i in tmp :
                data_list.append(i[0].text.strip())

            df = df.append(pd.Series(data_list,index=컬럼명_list), ignore_index=True)

            button = WebDriverWait(driver, 3).until(
                    lambda x: x.find_element_by_css_selector("img[src='/images/sub/tab_kaptinfo1_out.png']"))
            button.click()
            time.sleep(1)
            
            tmp_str_1 = df.tail(1)['법정동주소'].values[0].split(' ')
            tmp_str_2 = df.tail(1)['명칭(단지코드)'].values[0][:df.tail(1)['명칭(단지코드)'].values[0].rfind('(')]
            
            
            current = '%s %s %s 크롤링 성공' %(tmp_str_1[1], tmp_str_1[2], tmp_str_2)
            if current == before:
                err_flag = 1
                df = df.iloc[:-1]
                print('로봇인증 풀고 sgg: %d, emd: %d, danji: %d 부터 다시시작' %(sgg_index, emd_index, danji_index))
                break
            print(current + '------->(%d번째 단지)' % danji_index)
            before = current
            
            danji_index += 1
        
        emd_index += 1
        if err_flag == 1: #기능 테스트용 
            break
    sgg_index += 1
    if sgg_index == 6 or err_flag == 1: # 각 파트별로 5개 시군구 수집 후 종료 (A:6 , B:11, C:16, D:21, E:26)
        break
```
